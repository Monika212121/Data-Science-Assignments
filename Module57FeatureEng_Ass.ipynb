Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.

```python
# A1: To find Pearson Correlation coefficient

import pandas as pd

# Lets create a dataframe first
list= [[6, 85], [8, 90], [10, 95], [12, 98]]

df = pd.DataFrame(list, columns = ['No.of hours spend study(in hrs)', 'Marks scored in exam(out of 100)'], index = ['student1', 'student2', 'student3', 'student4'])

df.corr(method = 'pearson')
```

Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.

```python
# A2: To calculate Spearman's correlation coefficient

import pandas as pd

# Lets create a dataframe first
list= [[6, 7], [8, 9], [8.5, 9.5], [9, 9.8]]

df = pd.DataFrame(list, columns = ['No.of hours of sleep(in hrs)', 'Overall job satisfaction (scale 1-10)'], index = ['employee1', 'employee2', 'employee3', 'employee4'])

df.corr(method = 'spearman')
```

Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.

Q5. A survey was conducted to examine the relationship between age and preference for a particular
brand of soft drink. The survey results are shown below:

Age(Years), Soft drink preference
25 , Coke
42 , Pepsi
37 , Mountain dew
19 , Coke
31 , Pepsi
28 , Coke


```python
# A5: To calculate relationship between age and preference for a particular brand of soft drink.

import pandas as pd

# Lets create a dataframe first
list= [[25, 'Coke'], [42, 'Pepsi'], [37, 'Mountain Dew'], [19, 'Coke'], [31, 'Pepsi'], [28, 'Coke']]

df = pd.DataFrame(list, columns = ['Age of consumer(in years)', 'soft drink preferred'])

# Since there is categorical column data is present, we need to convert in into numerical data.
# We will achieve this by One Hot Encoding technique.

from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()

encoded_data = encoder.fit_transform(df[['soft drink preferred']]).toarray()

encoded_df = pd.DataFrame(encoded_data, columns = encoder.get_feature_names_out())

# Now integrate this encoded_df in the original df
new_df = pd.concat([df, encoded_df] ,axis = 1)

# Calcualte Pearson Correlation
pearson_corr = new_df.corr(method = 'pearson', numeric_only= True)

pearson_corr
```
