Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

A1. Here are five common plots in the Seaborn library, along with their uses:

1. Scatter Plot
Description: Displays data points based on two numerical variables.
Usage: Useful for identifying relationships or patterns between two continuous variables. It can reveal correlations, clusters, or outliers.
Example: Examining the relationship between advertising spending and sales revenue.
Code: sns.scatterplot(x="total_bill", y="tip", data=tips)

2. Bar Plot
Description: Visualizes categorical data with rectangular bars representing the value of each category.
Usage: Effective for comparing quantities across discrete categories. Itâ€™s commonly used to show counts, means, or sums.
Example: Showing average exam scores across different student groups.
Code: sns.barplot(x="day", y="total_bill", data=tips)

3. Box Plot
Description: A graphical representation of the distribution of data through quartiles.
Usage: Ideal for identifying the spread, skewness, and outliers in continuous data. It shows the median, quartiles, and potential outliers.
Example: Comparing income distributions across different age groups.
Code: sns.boxplot(x="day", y="total_bill", data=tips)

4. Heatmap
Description: A color-coded matrix that visualizes relationships between variables.
Usage: Commonly used to display correlations between variables or to visualize data with two categorical dimensions, such as a pivot table.
Example: Showing the correlation matrix of variables in a dataset.
Code: sns.heatmap(data.corr(), annot=True, cmap="coolwarm")

5. Histogram
Description: Visualizes the frequency distribution of a continuous variable by dividing data into intervals or bins.
Usage: Useful for understanding the distribution, skewness, and modality of data.
Example: Displaying the distribution of ages in a population dataset.
Code: sns.histplot(data=tips, x="total_bill", kde=True)
Each of these plots helps analyze data in different ways, providing insights into distributions, relationships, comparisons, and patterns within the data.


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.























```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "fmri" dataset
fmri_data = sns.load_dataset("fmri")

# Create the line plot with timepoint on x-axis and signal on y-axis, differentiated by event and region
plt.figure(figsize=(12, 6))
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri_data)
plt.title("FMRI Signal Over Time by Event and Region")
plt.xlabel("Timepoint")
plt.ylabel("Signal")
plt.show()
```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "titanic" dataset
titanic_data = sns.load_dataset("titanic")

# Set up the figure for two side-by-side box plots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Box plot for 'age' by 'pclass'
sns.boxplot(x="pclass", y="age", data=titanic_data, ax=axes[0])
axes[0].set_title("Age Distribution by Passenger Class")
axes[0].set_xlabel("Passenger Class")
axes[0].set_ylabel("Age")

# Box plot for 'fare' by 'pclass'
sns.boxplot(x="pclass", y="fare", data=titanic_data, ax=axes[1])
axes[1].set_title("Fare Distribution by Passenger Class")
axes[1].set_xlabel("Passenger Class")
axes[1].set_ylabel("Fare")

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "diamonds" dataset
diamonds_data = sns.load_dataset("diamonds")

# Plot a histogram for 'price' with 'cut' as hue
plt.figure(figsize=(12, 6))
sns.histplot(data=diamonds_data, x="price", hue="cut", multiple="stack", kde=True)
plt.title("Price Distribution of Diamonds by Cut Quality")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "iris" dataset
iris_data = sns.load_dataset("iris")

# Create the pair plot with hue for the "species" column
sns.pairplot(iris_data, hue="species", palette="viridis")
plt.suptitle("Pair Plot of Iris Dataset by Species", y=1.02)
plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "flights" dataset
flights_data = sns.load_dataset("flights")

# Create a pivot table for the heatmap with only three arguments
flights_pivot = flights_data.pivot(index="month", columns="year", values="passengers")

# Plot the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu", linewidths=.5)
plt.title("Flights Heatmap: Monthly Passengers per Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
